```python
import pandas as pd
import numpy as np

matches = pd.read_csv('../data/raw/matches.csv')
deliveries = pd.read_csv('../data/raw/deliveries.csv')

print('=== MATCHES ===')
print(f'Shape: {matches.shape}')
print(matches.dtypes)
```

```python
print('=== Missing Values — matches ===')
print(matches.isnull().sum())
print()
print('=== Missing Values — deliveries ===')
print(deliveries.isnull().sum())
```

```python
print('Seasons:', sorted(matches['season'].unique()))
print('Unique teams:', sorted(matches['team1'].unique()))
print('Venues:', matches['venue'].nunique())
print('Total matches:', len(matches))
print('Total deliveries:', len(deliveries))
```

## Initial Observations

- **matches.csv**: 1,095 records × 20 columns covering 17 seasons (2008–2024)
- **deliveries.csv**: 260,920 ball-by-ball records × 17 columns
- Season labels are inconsistent (`2007/08`, `2009/10`, `2020/21`) — requires standardisation
- `city` has 51 missing values (UAE bubble season)
- `method` has 1,074 nulls — normal matches where DLS was not applied
- Deliveries: `extras_type`, `player_dismissed`, `dismissal_kind`, `fielder` are null when the event did not occur — semantic nulls, not data errors
- Several franchise renames (`Delhi Daredevils` → `Delhi Capitals` etc.) cause double-counting risk
